```kotlin
data class ContingencyTable(
    val supportableCounts: LongArray,
    val unsupportableCounts: LongArray
) {

    val p = ChiSquareTest().chiSquareTest(arrayOf<LongArray>(supportableCounts, unsupportableCounts))

    val rowSums = longArrayOf(
        supportableCounts.sum(),
        unsupportableCounts.sum(),
    )
    val colSums = (0 until supportableCounts.size).map {
        supportableCounts[it] + unsupportableCounts[it]
    }

    val total = rowSums.sum()

    val residuals = Array(2) { i ->
        DoubleArray(supportableCounts.size) { j ->
            val expected = rowSums[i].toDouble() * colSums[j] / total
            ((supportableCounts.takeIf { i == 0 } ?: unsupportableCounts)[j] - expected) / sqrt(expected)
        }
    }

    val xi: Double = Array<DoubleArray>(2) { i ->
        DoubleArray(supportableCounts.size) { j ->
            val expected = rowSums[i].toDouble() * colSums[j] / total
            ((supportableCounts.takeIf { i == 0 } ?: unsupportableCounts)[j] - expected) / expected + 0.5
        }
    }.map { it.sum() }
        .sum()


    val cramersV = sqrt(xi / total.toDouble())
}
```

```kotlin
import org.apache.commons.math3.stat.inference.ChiSquareTest

val principles = listOf(
    "Следовала ли команда The Single Responsibility Principle",
    "Следовала ли команда The Open/Closed Principle",
    "Следовала ли команда The Liskov Substitution Principle",
    "Следовала ли команда The Interface Segregation Principle",
    "Следовала ли команда The Dependency Inversion Principle",
)

fun counts(principle: String): Pair<LongArray, LongArray> =
    (longArrayOf(
        supportable.getColumn(principle).count { it == "Никогда" }.toLong(),
        supportable.getColumn(principle).count { it == "Редко" }.toLong(),
        supportable.getColumn(principle).count { it == "Иногда" }.toLong(),
        supportable.getColumn(principle).count { it == "Часто" }.toLong(),
        supportable.getColumn(principle).count { it == "Всегда" }.toLong(),
    )
            to longArrayOf(
        unsupportable.getColumn(principle).count { it == "Никогда" }.toLong(),
        unsupportable.getColumn(principle).count { it == "Редко" }.toLong(),
        unsupportable.getColumn(principle).count { it == "Иногда" }.toLong(),
        unsupportable.getColumn(principle).count { it == "Часто" }.toLong(),
        unsupportable.getColumn(principle).count { it == "Всегда" }.toLong(),
    ))

val principlesContingencyTables = principles.associateWith() {
    val (supportableCounts, unsupportableCounts) = counts(it)
    ContingencyTable(supportableCounts, unsupportableCounts)
}
principlesContingencyTables.forEach { (p, tbl) ->
    println(
        "$p:\n" +
                "p= ${tbl.p}\n" +
                "${
                    tbl.residuals.joinToString("\n") { it.joinToString() { "%.2f".format(it) } }
                }\n" +
                "Cramer V: ${"%.2f".format(tbl.cramersV)} (total = ${tbl.total})"
    )
}
```

```kotlin
%use lets-plot
import org.jetbrains.letsPlot.*
import org.jetbrains.letsPlot.geom.*
import org.jetbrains.letsPlot.label.*

// Данные: standardized residuals для LSP (как пример)
val residuals = mapOf(
    "Да" to listOf(-1.62, -0.75, 0.83, 0.98, -0.17),
    "Нет" to listOf(2.26, 1.05, -1.15, -1.37, 0.23)
)

val categories = listOf("Никогда", "Редко", "Иногда", "Часто", "Всегда")

// Переводим в "длинный формат" для графика
val data = residuals.flatMap { (support, values) ->
    values.mapIndexed { idx, v ->
        mapOf(
            "Поддерживаемость" to support,
            "Ответ" to categories[idx],
            "Residual" to v
        )
    }
}

val df = data

// Рисуем heatmap
val p = ggplot(df) +
        geom_tile(
            aes(
                x = "Ответ",
                y = "Поддерживаемость",
                fill = "Residual"
            )
        ) +
        scale_fill_gradient2(
            low = "blue", mid = "white", high = "red", midpoint = 0.0
        ) +
        geom_text(aes("Ответ", "Поддерживаемость", label = "Residual"), color = "black") +
        ggsize(700, 300) +
        ggtitle("Стандартизированные остатки для LSP")

p
```
